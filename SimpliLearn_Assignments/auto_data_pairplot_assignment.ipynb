```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
auto_data = pd.read_csv("auto_data.csv")
auto_data.sample(10)
```

```python
auto_data.shape
```

```python
def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'

auto_data["origin"] = auto_data["origin"].apply(origin)
```

```python
sns.pairplot(auto_data[["mpg", "weight", "origin"]],size= 3)
```

```python
sns.pairplot(auto_data[["mpg", "weight", "origin"]],hue='origin',size= 3)
```
